```python
import imageio
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from skimage.transform import resize
from IPython.display import HTML
from IPython.display import Image
from IPython.display import clear_output
import warnings
warnings.filterwarnings("ignore")
```

```python
from demo import load_checkpoints
generator, kp_detector = load_checkpoints(config_path='config/vox-256.yaml', 
                            checkpoint_path='/content/drive/My Drive/first-order-motion-model/vox-cpk.pth.tar')
```

```python
from demo import make_animation
from skimage import img_as_ubyte
```

```python
def display(source, driving, generated1=None, generated2 = None):
    fig = plt.figure(figsize=(8 + 4 * (generated1 is not None) + 4 * (generated2 is not None), 6))
 
    ims = []
    for i in range(len(driving)):
        cols = [source]
        cols.append(driving[i])
        if generated1 is not None:
            cols.append(generated1[i])
        if generated2 is not None:
            cols.append(generated2[i])
        im = plt.imshow(np.concatenate(cols, axis=1), animated=True)
        plt.axis('off')
        ims.append([im])
 
    ani = animation.ArtistAnimation(fig, ims, interval=50, repeat_delay=1000)
    plt.close()
    return ani
```

```python
path_image='/content/drive/My Drive/first-order-motion-model/05.png' # Mona Lisa
path_video='/content/drive/My Drive/first-order-motion-model/10.mp4' # Video de Leonardo Dicaprio
source_image = imageio.imread(path_image)
driving_video = imageio.mimread(path_video)
 
#Resize image and video to 256x256
 
source_image = resize(source_image, (256, 256))[..., :3]
driving_video = [resize(frame, (256, 256))[..., :3] for frame in driving_video]
 
predictions = make_animation(source_image, driving_video, generator, kp_detector, relative=True)

predictions2 = make_animation(source_image, driving_video, generator, kp_detector, relative=False, adapt_movement_scale=True)
```

```python
HTML(display(source_image, driving_video, predictions, predictions2).to_html5_video())
```
